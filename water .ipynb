```python
# 1. Remove negative values (not possible in water datasets)
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    dataset = dataset[dataset[col] >= 0]

# 2. Remove zero or unrealistic water consumption
if "Water_Consumption_Liters" in dataset.columns:
    dataset = dataset[dataset["Water_Consumption_Liters"] > 0]

# 3. Remove invalid humidity values (valid 0–100)
if "Humidity_percent" in dataset.columns:
    dataset = dataset[(dataset["Humidity_percent"] >= 0) &
                      (dataset["Humidity_percent"] <= 100)]

# 4. Filter unrealistic building details
if "Floors" in dataset.columns:
    dataset = dataset[dataset["Floors"] <= 200]     # Max building height limit

if "Occupancy" in dataset.columns:
    dataset = dataset[dataset["Occupancy"] <= 50000]   # Avoid extreme wrong values

# 5. Remove duplicate rows
dataset = dataset.drop_duplicates()

# 6. Remove rows with >50% missing values
dataset = dataset.dropna(thresh=len(dataset.columns) * 0.5)

# 7. Reset index after cleaning
dataset = dataset.reset_index(drop=True)
```
